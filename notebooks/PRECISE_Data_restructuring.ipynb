## Unmapped Kenya villages
The workflow below was for identifying Kenya villages remaining unmapped after the Mombasa Precise Retreat

```python
ke_villages=pd.concat([anc_data[['county_V1', 'KE_CHU_V1', 'KE_Village_V1']].rename(columns={'county_V1': 'county', 'KE_CHU_V1': 'chu', 'KE_Village_V1': 'village'}),\
                       anc_data[['county_V2', 'KE_CHU_V2', 'KE_Village_V2']].rename(columns={'county_V2': 'county', 'KE_CHU_V2': 'chu', 'KE_Village_V2': 'village'}),\
                       anc_data[['county_delivery', 'KE_CHU_delivery', 'KE_Village_delivery']].rename(columns={'county_delivery': 'county', 'KE_CHU_delivery': 'chu', 'KE_Village_delivery': 'village'})])
ke_villages=(ke_villages.groupby(by=['chu', 'village'])
             .count()
             .dropna()
             .rename(columns={'county':'women'})
             .reset_index())
ke_villages['village']=ke_villages['village'].apply(lambda x: x.replace('-', ' '))
ke_villages['village']=ke_villages['village'].apply(lambda x: x.replace('_', ' '))
ke_villages['chu']=ke_villages['chu'].apply(lambda x: x.replace('-', ' '))
ke_villages['chu']=ke_villages['chu'].apply(lambda x: x.replace('_', ' '))
ke_villages
```

```python
ke_villages.to_csv('shapefiles/kenya_villages.csv')
```

```python
import geopandas as gpd
kenya=gpd.read_file('shapefiles/precise_villages.gpkg', layer='kenya')
```

```python
kenya
```

```python
kenya=kenya.drop_duplicates(subset=['chu', 'name'], keep='first')
kenya
```

```python
comb=kenya[['village_code', 'name', 'chu']].merge(ke_villages, left_on=['name', 'chu'], right_on=['village', 'chu'], how='right', validate='1:1')
comb[comb.village_code.isna()].sort_values(by='women', ascending=False).to_csv('shapefiles/kenya_unmapped.csv')
```
